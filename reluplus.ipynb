```python
import numpy as np
import matplotlib.pyplot as plt

# Definiowanie funkcji aktywacji
def linear(x):
    return x

def logistic(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x):
    return np.where(x > 0, x, x * 0.01)

def softmax(x):
    exp_x = np.exp(x - np.max(x))  # Stabilizacja numeryczna
    return exp_x / exp_x.sum(axis=0)

# Tworzenie zakresu wartości x
x = np.linspace(-10, 10, 400)

# Obliczanie wartości y dla każdej funkcji aktywacji
y_linear = linear(x)
y_logistic = logistic(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_leaky_relu = leaky_relu(x)

# Tworzenie wielowymiarowego zestawu danych dla funkcji softmax
x_softmax = np.linspace(-2, 2, 5)
y_softmax = np.array([softmax(np.array([xi, 0, 0, 0, 0]))[0] for xi in x])

# Rysowanie wykresów
plt.figure(figsize=(12, 8))

# Linear
plt.subplot(2, 3, 1)
plt.plot(x, y_linear, label="Linear", color="crimson")
plt.title("Linear")
plt.xlabel("Wejście")
plt.ylabel("Wyjście")
plt.grid(True)

# Logistic
plt.subplot(2, 3, 2)
plt.plot(x, y_logistic, label="Logistic", color="mediumvioletred")
plt.title("Logistic")
plt.xlabel("Wejście")
plt.ylabel("Wyjście")
plt.grid(True)

# Tanh
plt.subplot(2, 3, 3)
plt.plot(x, y_tanh, label="Tanh", color="darkorchid")
plt.title("Tanh")
plt.xlabel("Wejście")
plt.ylabel("Wyjście")
plt.grid(True)

# ReLU
plt.subplot(2, 3, 4)
plt.plot(x, y_relu, label="ReLU", color="deeppink")
plt.title("ReLU")
plt.xlabel("Wejście")
plt.ylabel("Wyjście")
plt.grid(True)

# Leaky ReLU
plt.subplot(2, 3, 5)
plt.plot(x, y_leaky_relu, label="Leaky ReLU", color="purple")
plt.title("Leaky ReLU")
plt.xlabel("Wejście")
plt.ylabel("Wyjście")
plt.grid(True)

# Softmax
plt.subplot(2, 3, 6)
plt.plot(x, y_softmax, label="Softmax", color="darkslateblue")
plt.title("Softmax")
plt.xlabel("Wejście")
plt.ylabel("Prawdopodobieństwo")
plt.grid(True)

plt.tight_layout()
plt.show()
```
